```python
def fit_one_over_distance_function(d, data):
    from scipy.optimize import leastsq
#     d = d * 1000
    guess_pho = 1
    optimize_func = lambda pho: 1/(pho * d) - data
    est_pho = leastsq(optimize_func, [guess_pho])[0]
    data_fit = 1/(est_pho * d)
    return est_pho, data_fit

def find_derivative_of_fitted_curve_at_d (d, est_pho):
    return (1. )/(est_pho * pow(d,2)) * 1000
```

We need to use this table to generate the fig2b plot. First we will group the data based on extracellular_electrode_distance, extracellular amplitude and frequency. Meaning that all the experiments with the same dextracellular amplitude, freq and distance of the electrode will be in the same group. Then we will compute the mean and std of the Vi, Vext and Vm amplitude and phase in each group. Then we plot these mean and std values as a function of frequencies. The function bellow, takes the subt_PCs and subt_PVs and retruns another table which has all the mean and std values for each group of data.

```python
def generate_subthreshold_data(table):
    subt_data = pd.DataFrame(columns=['ex_frequency', 'ex_amp(nA)', 'ex_el_distance(mu)',
                                      'mean_vi_amp(mV)', 'std_vi_amp', 'mean_vi_phase','std_vi_phase',
                                      'mean_vm_amp(mV)', 'std_vm_amp', 'mean_vm_phase','std_vm_phase',
                                      'mean_vext_amp(mV)', 'std_vext_amp', 'mean_vext_phase', 'std_vext_phase'])

    for (freq,ex_amp, distance),g in table.groupby(['ex_frequency', 'ex_amp(nA)', 'ex_el_distance(mu)']):
        subt_data = subt_data.append({'ex_frequency' : freq,
                                      'ex_el_distance(mu)' : distance,
                                      'ex_amp(nA)' : ex_amp, 
                                      'mean_vi_amp(mV)' : np.mean(g['vi_amp(mV)']),
                                      'mean_vi_phase' : np.mean(g['vi_phase']),
                                      'std_vi_amp' : np.std(g['vi_amp(mV)']),
                                      'std_vi_phase' : np.std(g['vi_phase']),                                      
                                      'mean_vm_amp(mV)' : np.mean(g['vm_amp(mV)']),
                                      'mean_vm_phase' : np.mean(g['vm_phase']),
                                      'std_vm_amp' : np.std(g['vm_amp(mV)']),                                      
                                      'std_vm_phase' : np.std(g['vm_phase']),
                                      'mean_vext_phase': np.mean(g['vext_phase']),
                                      'mean_vext_amp(mV)' : np.mean(g['vext_amp(mV)']),
                                      'std_vext_phase': np.std(g['vext_phase']),
                                      'std_vext_amp' : np.std(g['vext_amp(mV)'])} ,ignore_index = True)
    return subt_data
```

Using the above functions, we generate the subt_PCs_data and subt_PVs_data that contain the mean and std for each groups as the following:

```python
subt_PCs_data = generate_subthreshold_data(subt_PCs)
subt_PVs_data = generate_subthreshold_data(subt_PVs)
```

Now we are going to plot fig2b using the subt_PC_data and subt_PV_data tables. The following two functions will help us to do so:

```python
def plt_prop(xlabel, ylabel, xlim, ylim):
    from matplotlib import rc, rcParams
    rc('font', weight='medium')
    rc('axes', linewidth=1.5)
    fig, (ax) = plt.subplots(1, 1)
    fig.set_figheight(6)
    fig.set_figwidth(6)#     
    ax.set_xlabel(xlabel, fontsize=20, weight='bold')
    ax.set_ylabel(ylabel, fontsize=20, weight='bold' )
    ax.legend(fontsize=20)
    ax.tick_params(labelsize=20)
    ax.set_xlim(xlim[0], xlim[1])
    ax.set_ylim(ylim[0], ylim[1])
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    return ax
```

```python
def plot_fig2b(data, ex_amp, ex_el_distance, drop_freq_70, var_to_plot = "voltage" , title=None):
    t = data[(data['ex_amp(nA)'] == ex_amp) & (data['ex_el_distance(mu)']== ex_el_distance)]
    if drop_freq_70:
        ndx70 = t[t['ex_frequency']==70].index[0]
        t = t.drop(ndx70)
    if var_to_plot == "voltage":
        ax = plt_prop("Frequency (Hz)", "Amplitude(mV)", (0,101), (0,3))
        ax.set_title(title, size=20)
        ax.errorbar(t['ex_frequency'],t['mean_vi_amp(mV)'], t['std_vi_amp'], capsize=10, linewidth=4, color='navy', label = "Vi")
        ax.scatter(t['ex_frequency'],t['mean_vi_amp(mV)'], color='navy', label = "Vi", s=200)
        ax.errorbar(t['ex_frequency'],t['mean_vext_amp(mV)'], t['std_vext_amp'], capsize=10, linewidth=4, color='mediumvioletred', label = "Vext")
        ax.scatter(t['ex_frequency'],t['mean_vext_amp(mV)'], color='mediumvioletred', label = "Vext", s=200)
        ax.errorbar(t['ex_frequency'],t['mean_vm_amp(mV)'], t['std_vm_amp'], capsize=10, linewidth=4, color='mediumseagreen', label = "Vm")
        ax.scatter(t['ex_frequency'],t['mean_vm_amp(mV)'],  color='mediumseagreen', label = "Vm", s=200)
        ax.set_xticks([0, 50, 100])
        ax.set_yticks([0, 1, 2, 3])
    else:
        ax = plt_prop("Frequency (Hz)", "Phase(Degree)", (0,101), (0,3))
        ax.set_title(title, size=20)
        ax.errorbar(t['ex_frequency'],t['mean_vi_phase'], t['std_vi_phase'], capsize=10, linewidth=4, color='navy',label = "Vi")
        ax.scatter(t['ex_frequency'],t['mean_vi_phase'],  color='navy',label = "Vi", s=200)
        ax.errorbar(t['ex_frequency'],t['mean_vext_phase'], t['std_vext_phase'], capsize=10, linewidth=4, color='mediumvioletred',label = "Vext")
        ax.scatter(t['ex_frequency'],t['mean_vext_phase'], color='mediumvioletred',label = "Vext", s=200)
        ax.errorbar(t['ex_frequency'],t['mean_vm_phase'], t['std_vm_phase'], capsize=10, linewidth=4, color='mediumseagreen',label = "Vm")
        ax.scatter(t['ex_frequency'],t['mean_vm_phase'],  color='mediumseagreen',label = "Vm", s=200)
        ax.set_ylim(-50,370)
        ax.set_yticks([0, 180, 360])
        ax.set_xticks([0, 50, 100])
#         ax.set_yticks([0, 1, 2, 3])
#     plt.legend(fontsize=20)
    return plt.show()
```

```python
def generate_subthreshold_data_fig2c(table):
    subt_data = pd.DataFrame(columns=['<vm>', 'ex_amp(nA)', 'ex_el_distance(mu)',
                                      'mean_vi_amp(mV)', 'std_vi_amp', 'mean_vi_phase','std_vi_phase',
                                      'mean_vm_amp(mV)', 'std_vm_amp', 'mean_vm_phase','std_vm_phase',
                                      'mean_vext_amp(mV)', 'std_vext_amp', 'mean_vext_phase', 'std_vext_phase'])

    table['<vm>'] = pd.cut(table["avg_vm(mV)"], 7, include_lowest=True)
    for (vm,ex_amp, distance, freq),g in table.groupby(['<vm>', 'ex_amp(nA)', 'ex_el_distance(mu)', 'ex_frequency']):
        subt_data = subt_data.append({'<vm>' : vm.mid,
                                      'ex_el_distance(mu)' : distance,
                                      'ex_frequency' : freq,
                                      'ex_amp(nA)' : ex_amp, 
                                      'mean_vi_amp(mV)' : np.mean(g['vi_amp(mV)']),
                                      'mean_vi_phase' : np.mean(g['vi_phase']),
                                      'std_vi_amp' : np.std(g['vi_amp(mV)']),
                                      'std_vi_phase' : np.std(g['vi_phase']),                                      
                                      'mean_vm_amp(mV)' : np.mean(g['vm_amp(mV)']),
                                      'mean_vm_phase' : np.mean(g['vm_phase']),
                                      'std_vm_amp' : np.std(g['vm_amp(mV)']),                                      
                                      'std_vm_phase' : np.std(g['vm_phase']),
                                      'mean_vext_phase': np.mean(g['vext_phase']),
                                      'mean_vext_amp(mV)' : np.mean(g['vext_amp(mV)']),
                                      'std_vext_phase': np.std(g['vext_phase']),
                                      'std_vext_amp' : np.std(g['vext_amp(mV)'])} ,ignore_index = True)
    return subt_data
```

Here we write couple of functions to group the supra_PCs and supra_PVs tables as the follwoing: first we group the supra table by the frequency and extracellular amplitude. Then we gather all the corrected spike phases and use pycircstat package to compute the p_value, vector length, vector angle and ... and return a table with all these values. 

```python
def return_rose_data(table, ISI=False, n_sample=None):
    import random
    data = pd.DataFrame(columns=['ex_frequency', 'ex_amp(nA)', 'p_value', 'v_l', 'v_l_std', 'total_n_spikes', 
                                 'mean_firing_rate', 'std_firing_rate', 'v_angle', 'all_spike_phases'])
    
    for (freq,ex_amp),g in table.groupby(['ex_frequency', 'ex_amp(nA)']):
        if ISI:
            list1 = g['spike_phase_A_corrected'].sum()
            list2 = g['spike_ISI'].sum()
            ndx= [list2.index(x) for x in list2 if (x>freq -2) & (x<freq + 2)]
            all_spike_phases = [list1[x] for x in ndx]
        else:
            all_spike_phases = g['spike_phase_A_corrected'].sum()
        if n_sample: all_spike_phases = random.sample(all_spike_phases, n_sample)
        total_n_spikes = len(all_spike_phases)
        if (len(all_spike_phases) != 0):
            v_angle = pycircstat.descriptive.mean(np.array(all_spike_phases))
            v_length = pycircstat.descriptive.vector_strength(np.array(all_spike_phases)) 
            p_value = pycircstat.rayleigh(np.array(all_spike_phases))[0]
            data = data.append({'ex_frequency' : freq,
                                'mean_firing_rate': np.mean(g['firing_rate']),
                                'std_firing_rate': np.std(g['firing_rate']),
                                'all_spike_phases': all_spike_phases,
                                'ex_amp(nA)' : ex_amp, 
                                'p_value': p_value,
                                'v_l': v_length,
                                'v_l_std': pycircstat.descriptive.std(np.array(all_spike_phases)),
                                'total_n_spikes': len(all_spike_phases),
                                'v_angle': v_angle * 180 / np.pi} ,ignore_index = True)
    return data
```

```python
def plot_polar_histogram_grid(rose_data, amp_list , fq_list, m=20, n_sample=None):
    import random
    fig, axs = plt.subplots(len(fq_list), len(amp_list), figsize=(20, 25), facecolor='w', edgecolor='k', subplot_kw=dict(polar=True))
    fig.subplots_adjust(hspace = 1.2, wspace=.0001)
    if len(fq_list) or len(amp_list) > 1:
        axs = axs.ravel()
        
    for i, (fq,amp) in enumerate([[fq,amp] for fq in fq_list for amp in amp_list]):
        t = rose_data[(rose_data['ex_amp(nA)']==amp) & (rose_data['ex_frequency']==fq)]
        all_spikes = t['all_spike_phases'].sum()

        if len(all_spikes) == 0:
            print 'There were no spikes at amplitude {} and frequency {}.'.format(amp, fq)
            continue
        if n_sample: all_spikes = random.sample(all_spikes, n_sample)
        total_n_spikes = len(all_spikes)
        v_angle = pycircstat.descriptive.mean(np.array(all_spikes))
        v_length = pycircstat.descriptive.vector_strength(np.array(all_spikes))
        p_value = pycircstat.rayleigh(np.array(all_spikes))[0]
        Y, X = np.histogram([x * 180./np.pi for x in all_spikes], bins=15)
        Xp =(X[1:] + X[:-1]) / 2
        Xp = Xp * np.pi / 180
        bars = axs[i].bar(Xp, Y, width=0.35)
        thetaticks = np.arange(0,360,90)
        axs[i].set_thetagrids(thetaticks, frac=1.2, size=30)
        axs[i].set_yticks([])
        axs[i].set_ylim(0, m)
        axs[i].annotate('',xy=(v_angle, v_length  * 2* m), xytext=(v_angle,0), xycoords='data', arrowprops=dict(width=5, color='red'))
        axs[i].set_xlabel('Frequency: {}  amplitde: {}\nTotal number of Spikes: {}\nVector \
         length: {:0.3f}\nVector angle: {:0.3f}\nP-value: {:0.3E}'.format(fq, amp, total_n_spikes, v_length, 
                                                                         v_angle * 180./np.pi, p_value), size=15)        

    plt.tight_layout()    
    return axs
```
